```python
print('Median of units sold is',df['units_sold'].median())
print('Mean of units sold is',df['units_sold'].mean())
#The value_counts() function is used to get a Series containing counts of unique values.
df['units_sold'].value_counts()
# Series.value_counts(self, normalize=False, sort=True, ascending=False, bins=None, dropna=True)
df['units_sold'].value_counts(normalize=True) # relative frequency
df['units_sold'].value_counts(bins=10) # Rather than count values, group them into half-open bins, a convenience for pd.cut, only works with numeric data.
```
